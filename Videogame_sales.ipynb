```python
#I want to increase NA_Sales by 5
videogame.NA_Sales=videogame.NA_Sales.map(lambda x:x+5)
```

```python
videogame.head()
#now we can see in our dataset na_sales value has been increased by 5
```

```python
#I want to increase EU_Sales by 10

videogame.EU_Sales=videogame.EU_Sales.map(lambda x:x+10)
```

```python
videogame.head()
```

```python
#NA_Sales and EU_Sales both must be reduced by 5
#now I will use apply()
def sub(row):
  row.NA_Sales -=5
  row.EU_Sales -=5
  return row

videogame=videogame.apply(sub,axis=1)

```
